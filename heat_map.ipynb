```python
import pandas as pd
df=pd.read_csv("/content/drive/MyDrive/DBMS_project_final/Source Code/6-map/data.csv")
```

```python
import folium
from folium.plugins import HeatMap


# Clean the dataset by removing rows with missing values
df = df.dropna(subset=['latitude', 'longitude'])

# Create a map centered at the first crime location
map_center = [df['latitude'].iloc[0], df['longitude'].iloc[0]]
m = folium.Map(location=map_center, zoom_start=12)

# Create a heatmap using KDE technique
heat_data = df[['latitude', 'longitude']].values
m.add_child(HeatMap(heat_data, radius=15))

# Display the map
m
```

```python
data=df
import matplotlib.pyplot as plt
# Count the number of crimes in each category
crime_counts = df['Description'].value_counts()

# Plot a bar chart of the crime categories
plt.figure(figsize=(12,6))
plt.bar(crime_counts.index, crime_counts.values)
plt.xticks(rotation=90)
plt.title('Number of crimes by category')
plt.xlabel('Category')
plt.ylabel('Number of crimes')
plt.show()
```

```python
# Convert the Reported_Date column to a datetime object
df['Reported_Date'] = pd.to_datetime(df['Reported_Date'], format='%m/%d/%Y')

# Count the number of crimes per day
daily_counts = df.groupby('Reported_Date').size()

# Plot a line chart of the daily crime counts
plt.figure(figsize=(12,6))
plt.plot(daily_counts.index, daily_counts.values)
plt.title('Daily crime counts')
plt.xlabel('Date')
plt.ylabel('Number of crimes')
plt.show()
```
